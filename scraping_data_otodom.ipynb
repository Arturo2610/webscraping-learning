```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = 'https://www.olx.pl/motoryzacja/samochody/?search%5Bfilter_float_price:to%5D=10000&search%5Bfilter_enum_petrol%5D%5B0%5D=lpg'
r = requests.get(URL)
soup = BeautifulSoup(r.text, 'html')
```

```python
print(soup)
```

```python
results = soup.find_all('div', {'class': 'css-qfzx1y'})
```

```python
print(len(results))
```

```python
car_list = []
for item in results:
    car = {
        'name': item.find('h6', {'class': 'css-16v5mdi er34gjf0'}).text,
        'price': item.find('p', {'class': 'css-10b0gli er34gjf0'}).text.replace('zł', '').replace('do negocjacji', '').replace(' ','').strip(),
        'location_date': item.find('p', {'class': 'css-veheph er34gjf0'}).text,
        }
    car_list.append(car)
    
```

```python
print(car_list)
```

```python
df = pd.DataFrame(car_list)
df
```
