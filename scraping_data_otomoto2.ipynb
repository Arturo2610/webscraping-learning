```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def get_data(URL):
    r = requests.get(URL) # wyslanie requesta
    content = r.content.decode('utf-8') # dekodowanie bitow do tekstu
    soup = BeautifulSoup(r.text, 'html') #utworzenie obiektu soup
    results = soup.find_all('section', {'class': 'ooa-1rnjex3 ev7e6t817'}) # wyszukanie znacznikow ktore przchowuja tytuly
    car_list = []
    for item in results:
        car = {
            'name': item.find('h1', {'class': 'ev7e6t89 ooa-1xvnx1e er34gjf0'}).text,
            'price': item.find('h3', {'class': 'ev7e6t82 ooa-bz4efo er34gjf0'}).text.replace(' ','').strip(),
            'mileage': item.find('dd', {'data-parameter': 'mileage'}).text.replace('km', '').replace(' ','').strip(),
            'fuel_type': item.find('dd', {'data-parameter': 'fuel_type'}).text,
            'gearbox': item.find('dd', {'data-parameter': 'gearbox'}).text,
            'year': item.find('dd', {'data-parameter': 'year'}).text,
            'location': item.find('p', {'class': 'ooa-gmxnzj'}).text,
            # 'link': item.find('h1', {'class': 'ev7e6t89 ooa-1xvnx1e er34gjf0'})['href'],
            }
        car_list.append(car)
        
    return car_list
```

```python
def main():
    wszystkie_tytuly = []
    for page in range(1, 24): 
        URL = f'https://www.otomoto.pl/osobowe/lodzkie?search%5Bfilter_enum_damaged%5D=0&search%5Bfilter_float_price%3Ato%5D=10000&page={page}/'
        print('Pobieranie danych ze strony: ', URL)
        tytuly_ze_strony = get_data(URL)
        wszystkie_tytuly.extend(tytuly_ze_strony)
    global df
    df = pd.DataFrame(wszystkie_tytuly)
    df.to_csv(r'C:\Users\artur\Documents\Projects\scrapowanie-nauka\CSV\Scraping_otomoto2.csv', index = False)
```

```python
if __name__ == "__main__":
    main()
```
