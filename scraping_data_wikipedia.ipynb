```python
from bs4 import BeautifulSoup
import requests
```

```python
URL = 'https://en.wikipedia.org/wiki/List_of_World_Rally_Championship_Drivers%27_champions'
page = requests.get(URL)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[2]
```

```python
print(table)
```

```python
soup.find('table', ='sortable plainclass_rowheaders wikitable')
```

```python
soup.find_all('th', scope='col')
```

```python
titles = table.find_all('th', scope='col')
```

```python
titles
```

```python
table_titles = [title.text.strip() for title in titles]

print(table_titles)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(columns=table_titles)

df = df.rename(columns={'Points[a]' : 'Points'})

df
```

```python
table.find_all('tr')
```

```python
column_data = table.find_all('tr')
```

```python
print(column_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
    length = len(df)
    pd.concat([df, pd.DataFrame([individual_row_data])], ignore_index=True)
```
