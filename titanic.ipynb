```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('http://analityk.edu.pl/wp-content/uploads/2020/02/titanic.csv')
df.head()
```

```python
df = df[['pclass', 'survived', 'sex' , 'age']]
```

```python
df.isnull().sum()
```

```python
df.loc[:,'survived'][df['survived']==0] = 'No'
df.loc[:,'survived'][df['survived']==1] = 'Yes'
df.head()
```

```python
sns.catplot(data=df,
            x = 'survived',
            kind = 'count')
```

```python
sns.catplot(data=df,
            x = 'survived',
            kind = 'count',
            hue = 'pclass')
```

```python
sns.distplot(df[df['survived']=='Yes']['age'], bins = 20)
sns.distplot(df[df['survived']=='No']['age'], bins = 20)
plt.show()
```

```python
sns.catplot(data=df,
            x = 'survived',
            kind= 'count',
            hue='sex')
```
